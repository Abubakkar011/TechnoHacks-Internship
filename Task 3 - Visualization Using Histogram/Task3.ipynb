## Importing Required Libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Reading the CSV file

```python
df=pd.read_csv("C:\\Users\\ayesh\\Downloads\\Iris.csv")
df.head()
```

## Histogram of Sepal Length

```python
plt.hist(df['SepalLengthCm'],bins=20,color='purple')
plt.title('Sepal Length')
plt.grid()
```

## Histogram of Petal Length

```python
plt.hist(df['PetalLengthCm'],bins=20,color='purple')
plt.title('Petal Length')
plt.grid()
```

## Histogram of Sepal Width

```python
plt.hist(df['SepalWidthCm'],bins=20,color='purple')
plt.title('Sepal Width')
plt.grid()
```

## Histogram of Petal Width

```python
plt.title('Petal Width')
plt.hist(df['PetalWidthCm'],bins=20,color='purple')
plt.grid()
```
